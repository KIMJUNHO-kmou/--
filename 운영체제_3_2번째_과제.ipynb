```python
%%writefile sum.c
#include <stdio.h>

int main() {
    int sum = 0;
    for(int i=1; i<=10; i++) {
        sum += i;
    }
    return sum; // 종료 코드로 전달
}
```

```python
!gcc -o sum sum.c
```

```python
%%writefile prac3_7.c
#include <stdio.h>
#include <unistd.h>
#include <sys/types.h>
#include <sys/wait.h>

int main() {
    pid_t pid;
    int status;
    pid = fork();

    if(pid > 0) { // 부모 프로세스
        wait(&status);
        int ret = WEXITSTATUS(status);
        printf("1에서 10까지 합한 결과는 %d\n", ret);
        return 0;
    }
    else if(pid == 0) { // 자식 프로세스
        execl("./sum", "./sum", NULL);
    }
}
```

```python
!gcc -o prac3_7 prac3_7.c
```
